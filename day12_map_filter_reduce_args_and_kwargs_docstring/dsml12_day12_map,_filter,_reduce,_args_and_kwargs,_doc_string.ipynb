```python
# enumerate
# zip()
```

```python
data1 = ['ram', 'shyam', 'gita', 'rita']
data2 = [20,30,40,50]

for i in zip(data1, data2):
  print(i)
```

```python
for i, (name,age) in enumerate(zip(data1, data2)):
  print(i, name, age)
```

```python
for i, (name,age) in enumerate(zip(data1, data2), start = 1):
  print(i, name, age)
```

```python
# class task
# name, age, gender, salary
# name: ..., age: ..., gender: ..., salary: ...

name = ['ram', 'shyam', 'gita', 'rita']
age = [20,30,40,50]
gender = ['male', 'male', 'female', 'rita']
salary = [10000, 12000, 8000, 15000]

for i, (name, age, gender, salary) in enumerate(zip(name, age, gender, salary), start = 1):
  print(f"Name:{name}, Age:{age}, Gender:{gender}, Salary:{salary}")
```

```python
# reduce()

from functools import reduce # importing reduce() from the functools
mylist = [1,2,3,4,5]
def add(x, y):
  return x * y
```

```python
red_obj = reduce(add, mylist)
red_obj

# 1,2 = 2
# 2,3 = 6
# 6,4 = 24
# 24,5 = 120 : Output
```

```python
# classwork
# find the maximum number using reduce function

from functools import reduce
mylist = [4,9,2,8,1]
def maximum(x, y):
  if x > y:
    return x
  else:
    return y

result = reduce(maximum, mylist)
print("Maximum number:", result)
```

```python
# (2D) Area = l * w
# (3D) Volume = l * w * h

# Solve using the positional argument
def area(*args):
  if len(args) == 2:
    l,w = args
    print("2D Area =", l * w)
  elif len(args) == 3:
    l,w,h = args
    print("3D Volume =", l * w * h)
  else:
    print("Invalid number of argument")
```

```python
area(4,5)
area(4,5,6)
```

```python
# multiply
def mux(*args):
  result = 1
  for num in args:
    result *= num
  return result
```

```python
mux(2,3,4,5)
```

```python
mux(1,2,3,4,5,6,7)
```
