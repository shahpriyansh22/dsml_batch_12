```python
# voting machine
# class task

# user bata input age lini
# vote halna paucha ki paudaina
# condition: age 18+ and citizenship

age = int(input("Enter your age: "))
if age >= 18:
  citizenship = input("Do you have a citizenship? (yes/no)")
  if citizenship.lower() == "yes":
    print("You are eligible to vote")
  else:
    print("You are not eligible to vote")
else:
  print("You are under 18")
```

```python
# task
# age 60 bhanda mathi cha and membership cha bhani 30% discount
#                         and no membership bhani 20% discount
# "Not eligible for a senior discount"

age = int(input("Enter age: "))
if age >= 60:
  membership = input("Do you have a membership? (yes/no)")
  if membership.lower() == "yes":
    print("You get 30% discount")
  else:
    print("You get 20% discount only")
else:
  print("Not eligible for a senior discount")
```

```python
# SEE Grading System

marks = float(input("Enter your marks: "))

if marks >= 90:
    grade = "A+"
elif marks >= 80:
    grade = "A"
elif marks >= 70:
    grade = "B+"
elif marks >= 60:
    grade = "B"
elif marks >= 50:
    grade = "C+"
elif marks >= 40:
    grade = "C"
elif marks >= 35:
    grade = "D"
else:
    grade = "NG"

print("Your SEE Grade is:", grade)
```
