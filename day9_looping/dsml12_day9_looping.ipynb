```python
# break

count = 1
while count < 10:
  if count == 4:
    break
  print (count)
  count += 1
print("Loop ended")
```

```python
# continue

count = 0
while count < 10:
  count += 1
  if count == 3:
    continue
  print(count)
print("Loop ended")
```

```python
# pass

count = 0
while count < 10:
  count += 1
  if count == 3:
    pass
  print(count)
print("Loop ended")
```

```python
# class task
# Number Guessing Game

secret_number = 7

while True:
    guess = int(input("Enter your guess: "))

    if guess == secret_number:
        print("Correct Guess!")
        break
    else:
        print("Try Again")
```

```python
# ATM Menu System

balance = 1000

while True:
    print("\n----ATM Menu----")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Exit")

    choice = int(input("Choose option: "))

    # invalid choice
    if choice < 1 or choice > 3:
      print("Invalid choice. Try again.")
      continue

    # check balance
    if choice == 1:
      print("Your balance is:", balance)

    # deposit money
    elif choice == 2:
      amount = int(input("Enter deposit amount: "))
      balance += amount
      print("Deposit successful!")
      print("Updated balance is:", balance)

    # exit
    else:
      print("Thank you for using ATM!")
      break
```
