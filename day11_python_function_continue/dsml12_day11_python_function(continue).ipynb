```python
# Class task
# 1
def largest(a,b,c):
  if a > b and a > c:
    return a
  elif b > a and b > c:
    return b
  else:
    return c

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

# function call
result = largest(a,b,c)

print("Largest number:", result)
```

```python
# 2
def count_vowels(text):
  vowels = "aeiouAEIOU"
  count = 0

  for char in text:
    if char in vowels:
      count += 1
  return count
count_vowels("HELLoaJjsdfjf")
```
