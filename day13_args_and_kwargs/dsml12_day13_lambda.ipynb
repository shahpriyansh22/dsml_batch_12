```python
def sum(*args):
  total = 0
  for num in args:
    total += num
  return total
```

```python
sum(1,2,5)
```

```python
def add(*args):
  return max(args)
```

```python
def func(**kwargs):
  for k, val in kwargs.items():
    print(k, "=", val)

# s1 = 'python'
# s2 = 'is'
# s3 = 'awesome'
```

```python
func(s1 = 'python', s2 = 'is', s3 = 'awesome')
```

```python
def introduce(**kwargs):
  details = []
  for k, v in kwargs.items():
    print(k, v)
    details.append(k + ":" + str(v))
  return ",". join(details)
```

```python
introduce(Name = "Alice", Age = "25", City = "New York")
```

```python
# class task
# subject and student details
# position argument and details keyword argument

def details(**kwargs):
  data = []
  for k, v in kwargs.items():
    print(k, v)
    data.append(k + ":" + str(v))
  return ",". join(data)
```

```python
details(Subject = "DSML", StdName = "Priyansh", Age = "24")
```

```python
# Write a function that accepts products and their prices as keyword arguments and returns the total

# def product_details(*args, **kwargs):
#   print("Products:", args)
#   print("Prices:", kwargs)

def total_price(**products):
    total = 0
    for product, price in products.items():
        print(product, ":", price)
        total += price
    return total

total = total_price(Apple=100, Banana=140, Orange=90)

print("Total Price =", total)
```
