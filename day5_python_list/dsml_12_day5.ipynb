```python
lst3 = [[1,2,3],[4,5,6]]
lst4 = lst3.copy()
```

```python
lst3[1][1] = 100
print("original")
print(lst3)
print("copy")
print(lst4)
```

```python
import copy
lst5 = [[1,2,3],[4,5,6]]
lst6 = copy.deepcopy(lst5)
lst5[1][2] = 100
```

```python
print("original")
print(lst5)
print("copy")
print(lst6)
```
