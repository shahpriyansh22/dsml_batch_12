```python
# 2
def calculator(num1, num2, operator):
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        return num1 - num2
    elif operator == "*":
        return num1 * num2
    elif operator == "/":
        if num2 != 0:
            return num1 / num2
        else:
            return "Error! Division by zero is not allowed."
    else:
        return "Invalid operator!"

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

# Function call
result = calculator(num1, num2, operator)
print("Result =", result)
```

```python
# class task
def update_list():
    my_list = []

    for i in range(5):
        value = input("Enter a number: ")

        # Check if input is integer
        if value.isdigit():
            my_list.append(int(value))
            print("Integer added.")

        # Check if input is float
        elif "." in value:
            my_list.pop() if my_list else print("List is empty, nothing to remove.")
            print("Last element removed.")

        else:
            print("Invalid input.")

        # Display updated list
        print("Updated list:", my_list)


# Function call
update_list()
```

```python
# Another example
def lst_operation():
  number = []

  for i in range(5):
    value = input(f"Enter value {i + 1}: ")
    if "." in value:
      if len(number) > 0:
        removed = number.pop()
        print(f"Float detected. Removed: {removed}")
      else:
        print("Float detected, but the list is empty")
    else:
      number.append(int(value))
      print(f"Integer {value} added to the list")
    print("Updated List:", number)
```

```python
lst_operation()
```
